```python
# Load the dataset
data = pd.read_csv('ODI_Match_Data.csv')
data.iloc[0,:]
```

```python
#unique venues
data['venue'].unique()
```

```python
#describe the data
data.info
```

```python
#keeping rows of only first innings
data=data[data['innings']==1]
```

```python
import pandas as pd

# Select relevant columns from the dataset and make a copy of the DataFrame
modified_data = data[['match_id', 'batting_team', 'ball', 'striker', 'bowler', 'runs_off_bat', 'player_dismissed', 'venue', 'bowling_team']].copy()

# Display the DataFrame to verify the selection
modified_data

# Rename the 'striker' column to 'batsman' and 'runs_off_bat' to 'runs'
modified_data.rename(columns={
    'striker': 'batsman',
    'runs_off_bat': 'runs'
}, inplace=True)

# Convert 'player_dismissed' column: if a player is dismissed, mark it as 1, otherwise 0 (handling NaN values)
modified_data['player_dismissed'] = modified_data['player_dismissed'].apply(lambda x: 0 if pd.isna(x) else 1)

# Print the first 25 rows of the modified data to inspect the result
print(modified_data.head(25))
```

```python
modified_data
delivery_df=modified_data
```

```python
# Get a list of unique batting teams from the delivery DataFrame
delivery_df['batting_team'].unique()
```

```python
# Define a list of teams that you are specifically interested in
teams = [
    'Australia',
    'India',
    'Bangladesh',
    'New Zealand',
    'South Africa',
    'England',
    'West Indies',
    'Afghanistan',
    'Pakistan',
    'Sri Lanka'    
]
```

```python
# Filter the DataFrame to only include rows where the 'batting_team' is in the defined list of teams
delivery_df = delivery_df[delivery_df['batting_team'].isin(teams)]

# Similarly, filter the DataFrame to only include rows where the 'bowling_team' is in the list of teams
delivery_df = delivery_df[delivery_df['bowling_team'].isin(teams)]

# Display the filtered DataFrame to inspect the changes
delivery_df
```

```python
# Group by 'batsman' and calculate the total runs scored by each batsman, summing the 'runs' column
run_total = delivery_df.groupby('batsman')['runs'].agg('sum').reset_index(name='Runs')

# Sort the DataFrame based on total runs in descending order to get the top run-scorers
run_total = run_total.sort_values(by='Runs', ascending=False)

# Display the sorted DataFrame to see the run totals of all batsmen
run_total
```

```python
# Select specific columns from delivery_df to create the output DataFrame
output = delivery_df[['match_id','batting_team','bowling_team','ball','runs','player_dismissed','venue']]

output
```

```python
# Group by match_id and sum the runs for each match, then merge the total runs into the output DataFrame
total_df = output.groupby('match_id').sum()['runs'].reset_index()
output = output.merge(total_df, on=['match_id'], suffixes=('', '_total'))
print(output)

output
```

```python
# Split the ball column into 'over' and 'ball_no' to get over-wise and ball-wise information
output['over'] = output['ball'].apply(lambda x: str(x).split(".")[0])
output['ball_no'] = output['ball'].apply(lambda x: str(x).split(".")[1])
output
```

```python
# Calculate the total balls bowled by converting overs into balls and adding individual ball numbers
output['balls_bowled'] = (output['over'].astype('int')*6) + output['ball_no'].astype('int')
output
```

```python
# Convert 'runs' column to numeric to ensure proper calculations
output['runs'] = pd.to_numeric(output['runs'], errors='coerce')

# Calculate cumulative score (current_score) for each match as the game progresses using cumsum()
output['current_score'] = output.groupby('match_id')['runs'].cumsum()

# Calculate Current Run Rate (CRR) using the formula: (current_score * 6) / balls_bowled
output['crr'] = round((output['current_score'] * 6) / output['balls_bowled'], 2)
output
```

```python
# Calculate the number of wickets left by subtracting player_dismissed from 10
output['wickets_left'] = 10 - output['player_dismissed']
output
```

```python
# Calculate the total number of wickets fallen per match
wickets_per_match = output.groupby('match_id')['player_dismissed'].sum().reset_index()

# Rename the aggregated column for clarity
wickets_per_match = wickets_per_match.rename(columns={'player_dismissed': 'total_wickets'})

# Merge the total wickets information into the output DataFrame
output = output.merge(wickets_per_match, on=['match_id'])

print(output.head())
```

```python
# Drop the 'total_wickets' column as it's no longer needed
output.drop('total_wickets', axis=1, inplace=True)
output
```

```python
# Create final DataFrame with only the relevant columns for further analysis
final_df = output[['match_id','batting_team','bowling_team','runs','current_score','balls_bowled','wickets_left','crr','venue','runs_total']]

final_df
```

```python
# Calculate the number of balls left by subtracting balls_bowled from 300 (50 overs * 6 balls per over)
final_df['balls_left'] = 300 - final_df['balls_bowled']

# Ensure that if balls_left is less than 0, it's set to 0
final_df['balls_left'] = final_df['balls_left'].apply(lambda x: 0 if x < 0 else x)

# Drop the 'balls_bowled' column as it's not needed anymore
final_df.drop(columns=['balls_bowled'], inplace=True)

final_df
```

```python
# Group the final DataFrame by match_id
groups = final_df.groupby('match_id')
```

```python
# Convert current_score to numeric, handling any conversion errors
final_df['current_score'] = pd.to_numeric(final_df['current_score'], errors='coerce')

# Drop rows where 'current_score' is NaN to ensure clean data
final_df = final_df.dropna(subset=['current_score'])

# Get unique match IDs for iteration
match_ids = final_df['match_id'].unique()

# Initialize an empty list to store the sum of runs for the last 5 overs for each match
last_five = []

# Iterate through each match, calculate the rolling sum of runs for the last 5 overs (30 balls)
for id in match_ids:
    group = final_df[final_df['match_id'] == id]
    last_five.extend(group.rolling(window=30)['runs'].sum().tolist())

# Add the calculated last_five to the final DataFrame
final_df['last_five'] = last_five
```

```python
# Sort the DataFrame by match_id for better readability
final_df.sort_values(by='match_id', ascending=True, inplace=True)

# Drop any remaining NaN values
final_df.dropna(inplace=True)

# Get unique venues from the DataFrame
final_df['venue'].unique()
```

```python
# Filter venues where the count of games played is more than 1500
eligible_cities = final_df['venue'].value_counts()[final_df['venue'].value_counts() > 1500].index.tolist()
# Filter the DataFrame to only include games played in eligible venues
final_df = final_df[final_df['venue'].isin(eligible_cities)]

final_df
```

```python
# Separate features (X) and target (y) variables
X = final_df.drop(columns=['match_id', 'runs_total', 'current_score'])
y = final_df['runs_total']

# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Display the columns in the training set
X_train.columns
```

```python
# Define a column transformer to apply OneHotEncoding to categorical columns
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'venue'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor

import numpy as np

models = {
    # 'XGBRegressor': XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1),
    # 'LinearRegression': LinearRegression(),
    'RandomForestRegressor': RandomForestRegressor(n_estimators=100, random_state=1)
}
results={}
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Dictionary to store results
results = {}

# Loop through each model
for name, model in models.items():
    # Define a pipeline with transformer and model
    pipe = Pipeline(steps=[
        ('step1', trf),               # OneHotEncoding transformer
        ('step2', StandardScaler()),  # Standardize features
        ('step3', model)              # Model
    ])
    
    # Fit the model on the training data
    pipe.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = pipe.predict(X_test)
    
    # Calculate evaluation metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    # Calculate adjusted R2 score
    n = X_test.shape[0]  # Number of observations
    p = X_test.shape[1]  # Number of predictors
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    
    # Store the results
    results[name] = {
        'R2 Score': r2,
        'Mean Absolute Error': mae,
        'Root Mean Squared Error': rmse,
        'Adjusted R2 Score': adjusted_r2
    }

# Display the results
results
```
